```python
from google_play_scraper import app, reviews, Sort, reviews_all
# ambil data ulasan aplikasi webtoon
result = reviews_all(
    'com.naver.linewebtoon',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=1000
)
```

```python
import csv

with open('review_webtoon.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Ulasan'])
    for ulasan in result:
        writer.writerow([ulasan['content']])
```
